```python
import math
lngExtent = [115.9, 116.8]
latExtent = [39.5, 40.6]
lngGridSize = (lngExtent[1] - lngExtent[0]) / 100
latGridSize = (latExtent[1] - latExtent[0]) / 100
```

```python
def get_pos(t):
    x = np.floor((t.lng - lngExtent[0]) / lngGridSize)
    y = np.floor((t.lat - latExtent[0]) / latGridSize)
    return 100 * x + y
```

```python
porn = porn.assign(pos=get_pos)
```

```python
class2 = class2.assign(pos=get_pos)
```

```python
class3 = class3.assign(pos=get_pos)
```

```python
class4 = class4.assign(pos=get_pos)
class5 = class5.assign(pos=get_pos)
```

```python
t_ = [porn, class2, class3, class4, class5]
name = ['porn', 'class2', 'class3', 'class4', 'class5']
for j in range(5):
    t = t_[j]
    g = t.groupby('pos')
    cou = g.count()
    cou.reset_index(inplace=True)
    res = np.zeros(10000, dtype=np.int32)
    for i in range(cou.shape[0]):
        pos = int(cou.loc[i,'pos'])
        val = cou.loc[i, 'md5']
        if 0 <= pos < 10000:
            res[pos] = val
    res = res.reshape(100,100)
    with open(name[j]+'.txt', 'w') as f:    
        f.write(str(res.tolist()))
```
